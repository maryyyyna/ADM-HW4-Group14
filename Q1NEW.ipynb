```python
def permute(x, a, b, c):
    return (a*x + b) % c
```

```python
def createListOfPermutations(arr, permutations):
    listOfPermutations = []
    for j in range(30):
        l = {}
        for i,item in enumerate(arr):
            l[item] = permute(i, permutations[j][0], permutations[j][1], len(arr))

        listOfPermutations.append(l)

    return listOfPermutations
```

```python
def create_signature(df, listOfPermutations, nPermutations):
    signature_matrix=[]

    for row in df.iterrows():
        signature_k =[]
        for j in range(nPermutations):
            index=[listOfPermutations[j][row[1][s]] for s in range(6)]
            signature_k.append(min(index))
        
        signature_matrix.append(signature_k)

    return signature_matrix
```

```python
permutations = [(random.randint(0,len(final_list)), random.randint(0,len(final_list))) for _ in range(30)]
listOfPermutations = createListOfPermutations(final_list, permutations)
users['signature'] = create_signature(users, listOfPermutations, 5)
```
